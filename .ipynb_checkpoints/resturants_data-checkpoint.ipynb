```python
import requests
from bs4 import BeautifulSoup

# URL of the menu page
url = "https://villagesoulofindia.com/app/order/menu"

# Make a GET request to fetch the raw HTML content
response = requests.get(url)
if response.status_code == 200:
    print("Successfully fetched the webpage!")
else:
    print(f"Failed to fetch the webpage. Status code: {response.status_code}")
    exit()

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find all menu items
menu_items = soup.find_all('div', class_='card')

# Extract and display details
for item in menu_items:
    try:
        name = item.find('h5').get_text(strip=True)  # Item name
        description = item.find('span', style="font-size: 16px; text-align: center;").get_text(strip=True)  # Description
        price = item.find('div', class_='text-muted', style="text-align: center;").get_text(strip=True)  # Price
        
        print(f"Name: {name}")
        print(f"Description: {description}")
        print(f"Price: {price}")
        print("-" * 40)
    except AttributeError:
        # Skip items with missing data
        continue
```

```python
#####################################village-soul######################################
import requests
from bs4 import BeautifulSoup
import csv

# URL of the menu page
url = "https://villagesoulofindia.com/app/order/menu"

# Make a GET request to fetch the raw HTML content
response = requests.get(url)
if response.status_code == 200:
    print("Successfully fetched the webpage!")
else:
    print(f"Failed to fetch the webpage. Status code: {response.status_code}")
    exit()

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find all menu items
menu_items = soup.find_all('div', class_='card')

# Open a CSV file for writing
with open('menu_items.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # Write the header row
    writer.writerow(["Name", "Description", "Price"])

    # Extract details and write to the CSV
    for item in menu_items:
        try:
            name = item.find('h5').get_text(strip=True)  # Item name
            description = item.find('span', style="font-size: 16px; text-align: center;").get_text(strip=True)  # Description
            price = item.find('div', class_='text-muted', style="text-align: center;").get_text(strip=True)  # Price

            # Write the data row to CSV
            writer.writerow([name, description, price])
        except AttributeError:
            # Skip items with missing data
            continue

print("Data has been written to 'menu_items.csv'.")
```

```python
#####################################village-soul######################################
import requests
from bs4 import BeautifulSoup
import csv

# URL of the restaurant menu page
url = "https://villagesoulofindia.com/app/order/menu"

# Make a GET request to fetch the raw HTML content
response = requests.get(url)
if response.status_code == 200:
    print("Successfully fetched the webpage!")
else:
    print(f"Failed to fetch the webpage. Status code: {response.status_code}")
    exit()

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Initialize data storage
restaurant_name = "Village The Soul of India"
address = "48 W Old Country Rd, Hicksville, NY 11801"
open_close_times = "Mon-Sun: 11:30 AM - 10:00 PM"  # Hardcoded as it may not be available on the menu page

# Find all menu items
menu_items = soup.find_all('div', class_='card')

# Open a CSV file for writing
with open('village_restaurant_data.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # Write the header row
    writer.writerow(["Restaurant Name", "Address", "Open/Close Times", "Item Name", "Description", "Price"])

    # Extract menu item details and write to the CSV
    for item in menu_items:
        try:
            item_name = item.find('h5').get_text(strip=True)  # Menu item name
            description = item.find('span', style="font-size: 16px; text-align: center;").get_text(strip=True)  # Description
            price = item.find('div', class_='text-muted', style="text-align: center;").get_text(strip=True)  # Price

            # Write data for each menu item
            writer.writerow([restaurant_name, address, open_close_times, item_name, description, price])
        except AttributeError:
            # Skip items with missing data
            continue

print("All data has been written to 'village_restaurant_data.csv'.")
```

```python
#####################################hennai-dosas######################################
import requests
from bs4 import BeautifulSoup
import csv

# URL of the restaurant menu page
url = "https://www.yelp.com/menu/chennai-dosas-hicksville"

# Make a GET request to fetch the raw HTML content
response = requests.get(url)
if response.status_code == 200:
    print("Successfully fetched the webpage!")
else:
    print(f"Failed to fetch the webpage. Status code: {response.status_code}")
    exit()

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Initialize data storage
restaurant_name = "Chennai Dosas"
address = "128 S Broadway, Hicksville, NY 11801, United States"
open_close_times = "Mon-Sun: 11:30 am to 10:30 pm"  # Hardcoded as it may not be available on the menu page

# Find all menu items
menu_items = soup.find_all('div', class_="arrange")

# Initialize a set to keep track of item names and avoid duplicates
menu_data = set()

# Open a CSV file for writing
with open('chennai_dosas_data.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # Write the header row
    writer.writerow(["Restaurant Name", "Address", "Open/Close Times", "Item Name", "Description", "Price"])

    # Extract menu item details and write to the CSV
    for item in menu_items:
        try:
            item_name = item.find('h4').get_text(strip=True)  # Menu item name
            description = item.find('p', class_="menu-item-details-description").get_text(strip=True)  # Description
            price = item.find('li', class_="menu-item-price-amount").get_text(strip=True)  # Price

            # Only write the item if it is not already in the set (to avoid duplicates)
            if item_name not in menu_data:
                menu_data.add(item_name)
                # Write data for each menu item
                writer.writerow([restaurant_name, address, open_close_times, item_name, description, price])
        except AttributeError:
            # Skip items with missing data
            continue

print("All data has been written to 'chennai_dosas_data.csv'.")
```
